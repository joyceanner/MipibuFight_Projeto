# Preparando o Ambiente

## Carregando planilha

```python
import pandas as pd
sheet_id = '1gn2AoLOsYGtTpr5S0Kxl8l8_iXMg2uLBBoi72NKnFLg'
url = f'https://docs.google.com/spreadsheets/d/{sheet_id}/gviz/tq?tqx=out:csv&sheet'
dados = pd.read_csv(url)

dados
```

## Substituindo valores nulos e excluindo redundâncias

```python
dados.isnull().sum()
```

```python
dados = dados.fillna('0')
dados.isnull().sum()
```

```python
dados = dados.replace('MMA ', 'MMA')
dados['MODALIDADE'].unique()
mma = ['MMA']
dados.query('@mma in MODALIDADE')
lista_mma = dados.query('@mma in MODALIDADE')
```

```python
dados = dados.astype({
    'IDADE' :'int',
    'LUTAS': 'int',
    'OUTRAS LUTAS': 'int',
    'VITORIAS':'int',
    'DERROTAS': 'int',
})
```

# Lista com todos os competidores menores de idade

```python
competidores_menores_idade = dados.query('IDADE < 18.0').filter(['NOME' , 'IDADE', 'EQUIPE', 'CIDADE'])
competidores_menores_idade
```

# MMA

## Menores de 18 anos

```python
selecao_mma_menores_18 = lista_mma['IDADE'] < 18
lista_mma[selecao_mma_menores_18]
```

## Entre 18 e 25 anos

```python
mma_maior_18 = lista_mma['IDADE'] >= 18
mma_menor_25= lista_mma['IDADE'] <= 25

mma_entre_18_e_25 = (mma_maior_18) & (mma_menor_25)
lista_mma[mma_entre_18_e_25]
```

## Maiores de 25 anos

```python
mma_maior_25 = lista_mma['IDADE'] >= 26
lista_mma[mma_maior_25]
```
